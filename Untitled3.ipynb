```python
def load_flights(file):
        flights = []
        with open(file, "r") as fl:
            fl_reader = csv.reader(fl)
            for flight in fl_reader:
                flights.append(flight)
        keys = flights.pop(0)
        flights.sort(key=lambda x: x[3])
        return [list(flights), keys]
```

```python
f = load_flights("examples/example1.csv")
```

```python
def create_graph(data_list):
    ret_dict = {}
    for i in data_list:
        if i[1] not in ret_dict:
            ret_dict[i[1]] = []
        if i[2] not in ret_dict[i[1]]:
            ret_dict[i[1]].append(i[2])
    return ret_dict
```

```python
def find_connection(start, end, time = 0):
    if time == 0:
        ret_flights = []
        for flight in f[0]:
            if flight[1] == str(start) and flight[2] == end:
                ret_flights.append(flight)
        return ret_flights  
    else:
        time = return_datetime_format(time)
        ret_flights = []
        for flight in f[0]:
            if flight[1] == str(start) and flight[2] == end:
                if time + timedelta(seconds = 3600) <= return_datetime_format(flight[3]) <= time + timedelta(seconds = 21600):
                    ret_flights.append(flight)
        return ret_flights
```

```python
def find_flights(start, end, graph, path, all_path, temp_flights, all_flights):
        
        path.append(start)
        
        if (start == end):
            all_path.append(list(path))
            if min([int(float(a[7])) for a in f[0]]) >= 0:
                all_flights.append(return_right_format(temp_flights))   
        else:
            for i in graph[start]:
                if (i not in path):
                    if temp_flights == []:
                        # at the beginning 
                        flights = find_connection(start, i)
                    else:
                        # using layover rules
                        flights = find_connection(start, i, temp_flights[-1][4]) 
                    if flights != []:
                        for flight in flights:
                            temp_flights.append(list(flight))
                            find_flights(i, end, graph, path, all_path, temp_flights, all_flights)
                            temp_flights.pop()
        path.pop()
```

```python
def _find_flights(self, start, end, graph, path, all_path, df, all_df, ret, ret_time):
        
        path.append(start)

        if (start == end):
            # direct flight
            if ret == False:
                all_path.append(list(path))
                if df["bags_allowed"].min() >= self.bags:
                    all_df.append(self._return_dict_format(df, self.destination))   
            else:
                # return flight 
                all_path.append(list(path))
                path2 = []
                ret = False
                ret_time = True
                self._find_flights(end, self.origin, graph, path2, all_path, df, all_df, ret, ret_time)  
        else:
            for i in graph[start]:
                if (i not in path):
                    if df.empty == True:
                        # at the beginning 
                        flights = self._find_connection(start, i)
                    else:
                        # at the beginning of the way back 
                        if ret_time == True:
                            flights = self._find_connection(start, i, str(df.iloc[df.shape[0] - 1]["arrival"]), True)
                        else:
                            # using layover rules
                            flights = self._find_connection(start, i, str(df.iloc[df.shape[0] - 1]["arrival"]))                     
                    if flights.empty == False:
                        for flight in range(flights.shape[0]):
                            df = df.append(flights.iloc[flight])
                            self._find_flights(i, end, graph, path, all_path, df, all_df, ret, False)
                            df = df.drop(df.index[df.shape[0] - 1])
        path.pop()
    
```

```python
bags = 1
def return_right_format(lst):
    ret_dict = {"flights": []}
    for flight in lst:
        dictt = {}
        i = 0
        for key in f[1]:
            dictt[key] = flight[i]
            i += 1
        ret_dict["flights"].append(dictt)
        ret_dict["bags_allowed"] = min([int(a[7]) for a in lst])
        ret_dict["bags_count"] = bags
        ret_dict["destination"] = lst[-1][2]
        ret_dict["origin"] = lst[0][1]
        ret_dict["total_price"] = sum([int(float(a[5])) for a in lst]) + bags * sum([int(float(a[6])) for a in lst])
        ret_dict["travel_time"] = parse_time((return_datetime_format(lst[-1][4]) 
                                              - return_datetime_format(lst[0][3])).total_seconds())
    return ret_dict
```

```python
def return_datetime_format(s):
        return datetime.strptime(s, "%Y-%m-%dT%H:%M:%S")
```

```python
def parse_time(s):
        hours = int(s / 3600)
        minutes = int((s - hours * 3600) / 60)
        seconds = int(s - hours * 3600 - minutes * 60)
        if hours < 10:
            hours = str(0) + str(hours)
        if minutes < 10:
            minutes = str(0) + str(minutes)
        if seconds < 10:
            seconds = str(0) + str(seconds)
        return  str(hours) + ":" + str(minutes) + ":" + str(seconds)
```
